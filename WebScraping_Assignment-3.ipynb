### `2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv.`

### Details to be scraped are:
- "Brand Name"
- "Name of the Product"
- "Rating"
- "No. of Ratings"
- "Price"
- "Return/Exchange"
- "Expected Delivery"
- "Availability"
- "Other Details"
- “Product URL”.

```python
# creating empty lists:
Brand=[]
Product=[]
Rating=[]
N_Rating=[]
Price=[]
Return=[]
Delivery=[]
Availability=[]
Details=[]
URL=[]
```

```python
url = driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
```

```python
for z in url:
 try: 
     URL.append(z.get_attribute("href"))
 except:
    URL.append('-')
```

```python
import time 


i=2

while(i>0): 
    nextbtn = driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
    nextbtn.click()
    time.sleep(3)
    url= driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for u in url:
          URL.append(u.get_attribute("href"))
    i=i-1
```

```python
for z in URL:
    driver.get(z)
    time.sleep(2)
    try:
        brand = driver.find_element_by_xpath("//td[@class='a-size-base prodDetAttrValue']")
        Brand.append(brand.text)
    except:
        Brand.append('-')
    try:
        product = driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        Product.append(product.text)
    except:
        Product.append('-')
    try:
        rating = driver.find_element_by_xpath("//div[@id='averageCustomerReviews']//span[@class='a-icon-alt']")
        Rating.append(rating.get_attribute("innerHTML"))
    except NoSuchElementException:
        Rating.append('-')
    try:
        number = driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        N_Rating.append(number.get_attribute("innerHTML"))
    except:
        N_Rating.append('-')
    try:
        price = driver.find_element_by_xpath("//span[@id='priceblock_ourprice']")
        Price.append(price.text)
    except:
        Price.append('-')
    
    try:
        returnn = driver.find_element_by_xpath("//div[@class='a-accordion-row-a11y']//h5")
        Return.append(returnn.text.replace('\n',' ').replace('   ',''))
    except:
        Return.append('-')
    
    try:
        delivery = driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']//b")
        Delivery.append(delivery.text)
    except:
        Delivery.append('-')
    try:
        availability = driver.find_element_by_xpath("//div[@id='availability']//span")
        Availability.append(availability.text)
    except:
        Availability.append('-')
    try:
        detail = driver.find_element_by_xpath("//div[@id='feature-bullets']")
        Details.append(detail.text.replace('\nShow More','.').replace('\n',''))
    except:
        Details.append('-')
```

```python
Search_result=pd.DataFrame({})
Search_result['Brand Name']=Brand
Search_result['Name of the Product']=Product
Search_result['Rating']=Rating
Search_result['No. of Ratings']=N_Rating
Search_result['Price']=Price
Search_result['Return/Exchange']=Return
Search_result['Expected Delivery']=Delivery
Search_result['Availability']=Availability
Search_result['Other Details']=Details
Search_result['Product URL']=URL
Search_result
```

```python
# Saving as .csv file:

Search_result.to_csv('Laptopscrape.csv') 
```
