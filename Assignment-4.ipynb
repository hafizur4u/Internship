### 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
    url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/

You need to find following details:

- Rank
- Name
- Artist
- Upload date
- Views

```python
#importing all required libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import time
from bs4 import BeautifulSoup
import requests
```

```python
#connecting web driver
driver= webdriver. Chrome(r'C:\Users\hifzu\Downloads\chromedriver.exe')
```

```python
link= " https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(link)
```

```python
Rank = []
Name = []
Artist = []
Views = []
Date = []
```

```python
for i in range(1,31):
    rank=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[1]")
    try:
        Rank.append(rank.text)
    except NoSuchElementException:
        Rank.append("-")
        
       
for i in range(1,31):
    name=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[2]")
    try:
        Name.append(name.text.strip().split('"')[1]) #appending only the names of the videos
    except NoSuchElementException:
        Name.append("-")
 
for i in range(1,31):
    artist=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[3]")
    try:
        Artist.append(artist.text)
    except NoSuchElementException:
        Artist.append("-")
        
for i in range(1,31):
    views=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[4]")
    try:
        Views.append(views.text)
    except NoSuchElementException:
        Views.append("-")
        
        
for i in range(1,31):
    date=driver.find_element_by_xpath(f"//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr[{i}]/td[5]")
    try:
        Date.append(date.text)
    except NoSuchElementException:
        Date.append("-")
```

```python
Top_Youtuube_video=pd.DataFrame({})
Top_Youtuube_video['Rank']=Rank
Top_Youtuube_video['Name']=Name
Top_Youtuube_video['Artist']=Artist
Top_Youtuube_video['Views']=Views
Top_Youtuube_video['Date']=Date
Top_Youtuube_video
```

```python
#Will keep scrolling down the webpage until it cannot scroll no more
last_height = driver.execute_script('return document.body.scrollHeight')
while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    time.sleep(2)
    new_height = driver.execute_script('return document.body.scrollHeight')
    try:
        driver.find_element_by_xpath('/html/body/div[4]/div/div/div[2]/section/button').click()
        time.sleep(2)
    except:
        pass
    if new_height == last_height:
        break
    last_height = new_height
```

```python
Match = []
Series = []
Place = []
Date = []
Time = []
```

```python
match = driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
series = driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
place = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']//span")
date = driver.find_elements_by_xpath("//div[@class='fixture__full-date']")
time = driver.find_elements_by_xpath("//span[@class='fixture__time']")
```

```python
for m in match:
    try:
        Match.append(m.text)
    except NoSuchElementException:
        Match.append("-")
for s in series:
    try:
        Series.append(s.text)
    except NoSuchElementException:
        Series.append("-")
        
for p in place:
    try:
        Place.append(p.text)
    except NoSuchElementException:
        Place.append("-")
        
for d in date:
    try:
        Date.append(d.text)
    except NoSuchElementException:
        Date.append("-")
for t in time:
    try:
        Time.append(t.text) 
    except NoSuchElementException:
        Time.append("-")
```

```python
Cricket=pd.DataFrame({})
Cricket['Match']=Match
Cricket['Series']=Series
Cricket['Place']=Place
Cricket['Date']=Date
Cricket['Time']=Time
Cricket
```
